  Convert this format to ["Action","Adventure","Fantasy","Science Friction"] 

```python
 import ast  #A module which convert string of list to list 
#ast.literal_eval is function in it 
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

From cast we will fetch only top 3 fro m dictionary  and usme se bhi hum sirf name nikalenge of actors 

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L 
```

```python
movies['cast'].apply(convert3)
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(10)
```

[Sam Worthington to  SamWorthington ]

```python
def collapse(L):# function to remove spaces 
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x)) # coverts tags list to string 
new.head()
```

```python
new['tags'][0]
```

```python
new['tags']=new['tags'].apply(lambda x:x.lower())
new['tags'][0]
```

Use stemming 

[loved,loving,love] to [love,love,love] 

```python
pip install nltk
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(word):
    y=[]
    for i in word.split():
        y.append(ps.stem(i))
    return " ".join(y) 
```

```python
new['tags']=new['tags'].apply(stem)
```

Text Vectorization 


```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

4806-Movies and 5000 words, get all each movie in vector form 

```python
vector[0] #of each movie , no represent the frequency  
```

```python
cv.get_feature_names_out()# names of 5000 words 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[0]# of first movi  
```

```python
new[new['title'] == 'The Lego Movie'].index[0]# giving index of the given title of first row 
```

```python
list(enumerate(similarity[0]))# we have create index of each distances 
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
        
    
```

```python
recommend("The Dark Knight Rises")
```
